```python
import numpy as np
import tensorflow as tf
```

```python
from tensorflow.keras.layers import Embedding,SimpleRNN,Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.datasets import imdb
```

```python
(x_train,y_train),(x_test,y_test)=imdb.load_data(num_words=10000)
```

```python
max_len=500
```

```python
x_train=sequence.pad_sequences(x_train,maxlen=max_len)
x_test=sequence.pad_sequences(x_test,maxlen=max_len)
```

```python
x_train[0]
```

```python
model=Sequential()
model.add(Embedding(input_dim=10000,output_dim=128,mask_zero=True))
model.add(SimpleRNN(128,activation="relu"))
model.add(Dense(1,"sigmoid"))
```

```python
model.compile(optimizer='adam',loss="binary_crossentropy",metrics=["accuracy"])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop=EarlyStopping(monitor="val_loss",patience=5,restore_best_weights=True)
```

```python
history=model.fit(x_train,y_train,validation_split=0.2,batch_size=32,epochs=15,callbacks=[early_stop])
```

```python
model.save("simple_rnn_imdb.h5")
```

```python
from tensorflow.keras.models import load_model
```

```python
model=load_model("simple_rnn_imdb.h5")
```

```python
word_index=imdb.get_word_index()
```

```python
def decode_review(encoded_review):
  return " ".joint([imdb_word_ind_rev.get(i-3,"?") for i in encoded_review])

def process_text(text):
  words=text.lower().split()
  encoded_review=[[word_index.get(word,2)+3 for word in words]]
  padded_review=sequence.pad_sequences(encoded_review,maxlen=500)
  return padded_review
```

```python
def predict_sentiment(review):
  processed_input=process_text(review)
  prediction=model.predict(processed_input)
  sentiment="positive" if prediction[0][0]>0.5 else "negative"
  return sentiment,prediction[0][0]
```

```python
example_review="The movie was fantastic! the acting was good and the plot was thrilling"
```
